```python
data = data[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
x = data['Annual Income (k$)']
y = data['Spending Score (1-100)']
plt.scatter(x, y, color='red')
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

WCSS -> within clusters sum of squares

```python
#finding WCSS value for diffrent number of clusters

wcss = []

for i in range(1,11):
  kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
  kmeans.fit(data)
  wcss.append(kmeans.inertia_)
```

```python
#plotting the elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum no. of clusters will be 5

```python
#training the K-means clustering model
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)

#return a label for each data point
y_kmeans = kmeans.fit_predict(data)
y_kmeans
```

Visualizing all the clusters

```python
plt.scatter(x, y, c=y_kmeans, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=300, alpha=1)
plt.show()
```
