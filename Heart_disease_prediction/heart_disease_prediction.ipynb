```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data = pd.read_csv('heart_disease_data.csv')
```

```python
heart_data.head()
#age #sex #chest-pain-type #trestbps=resting-blood-pressure #chol=cholestrol #fbs=fasting-blood-sugar #restecg=resting-electrocardiographic results #thalach=maximum-heartrate-achieved #exang=exercise-included-angina #oldpeak= ST depression induced by exercise relative to rest #slope=the slope of the peak exercise ST segment #ca=no. of majot vessels #thal:3=normal, 6=fixed, 7=reversable defect #target: 0 = no disease, 1 = disease positive
```

```python
heart_data.describe()
```

```python
heart_data.isnull().sum()
```

```python
#checking the distribution of target variable
heart_data['target'].value_counts() #1=defective heart, 0=healthy heart
```

```python
#splitting the features and target
features = heart_data.drop(columns='target', axis=1)
target = heart_data['target']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.2, stratify=target, random_state=2) #stratify = it equally distributes the 0s and 1s in the variables. if we do not include stratify then there is  a possibility that all the 0s of the target gets in one variable(X_test) and all the 1s get in one variable etc
```

```python
#model training
model = LogisticRegression()
```

```python
#training the model with traning data
model.fit(X_train, Y_train)
```

```python
#doing this just to check the diff between train-prediction and test-prediction(if the diff is large then it is a problem, we should not move further)
#model evalution
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print('Accuracy score on Test Data : ', training_data_accuracy)
```

```python
#doing this just to check the diff between train-prediction and test-prediction(if the diff is large then it is a problem, we should not move further)
#accuracy on testing data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy score on Test Data : ', test_data_accuracy)
```

```python
#building a predictive system

first_input_data = (41,0,1,130,204,0,0,172,0,1.4,2,0,2) #should show heart disease = positive
second_input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2) #should show heart disease = negative

#change the input data to a numpy array
input_data_as_numpy_array = np.asarray(second_input_data)

#reshape the numpy array as we are predicting for only one instance(if we not do this then our model with predict for all the 80% data i.e 242 values)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if prediction == 0:
  print('The person does not have a heart disease')
else:
  print('The person has a heart disease')
```
